```python
import pandas as pd
```

```python
# import rosters
for team in ['TOR']:
    filename = team+'_roster_2019.csv'
    roster = pd.read_csv(filename)

    filename = team+'_per_game_2019.csv'
    # import per game stats
    game = pd.read_csv(filename)

    # merge two dataframes
    df = game.merge(roster,on='Player')
    df['Player'] = df['Player'].apply(lambda x: x.split("\\")[0])
    df['Team'] = team.lower()
```

```python
team_stats = pd.read_csv('team_stats_2019.csv',header=1)
```

```python
team_stats['Team'] = team_stats['Team'].apply(lambda x: x.strip('*'))
team_stats['Team'] = team_stats['Team'].apply(lambda x: x.lower())
```
